```python
# -*- coding: utf-8 -*-
# A nurse can either apply ('applied') for a job or skip it ('skipped').
# Than the manager will choose someone and send an offer ('confirmed') or reject the application of a nurse ('rejected').
# And now the nurse can choose to accept the job ('deal') or decline it ('rejectedByNurse').

import pandas as pd

file = './applications.csv'
dfapp = pd.read_csv(file, header =[0], skipinitialspace=True, encoding='utf-8')
# dfapp = pd.read_csv(file, header =[0], encoding='utf-8', parse_dates=['offeredDate'])
print(dfapp.shape)
print(list(dfapp))
print(dfapp.head())

file = './jobs.csv'
dfjob = pd.read_csv(file, header =[0], skipinitialspace=True, encoding='utf-8', parse_dates=['date', 'created'])
print(dfjob.shape)
print(list(dfjob))
dfjob.head()
```

```python
dfapp = dfapp[dfapp.status != 'skipped']
dfapp.drop(['id','staffId', 'jobDate','offeredDate','status', 'active'], axis=1, inplace=True)
dfapp = dfapp.groupby(['jobId']).size().reset_index(name='counts')
print(dfapp.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

print(dfapp.counts.unique())
print(dfapp.counts.describe())
plt.subplot(1, 2, 1)
(dfapp['counts']).plot.hist(bins=50, figsize=(12, 6), edgecolor = 'white', range = [0, 13])
plt.xlabel('nurses per job', fontsize=12)
plt.title('Nurses Distribution', fontsize=12)
```

```python
import datetime

#with more data might want to keep/transform date for month pattern
#title is unique. if not, keep it/dummy
#skills is text, not used right now

lastday = datetime.datetime(2019, 3, 5)
dfjob = dfjob[dfjob.date <= lastday]

fn = lambda row: (row.date - row.created).days
col = dfjob.apply(fn, axis=1)
dfjob = dfjob.assign(delay=col.values)

fn = lambda row: row.created.isoweekday()
col = dfjob.apply(fn, axis=1)
dfjob = dfjob.assign(created_day=col.values)

fn = lambda row: row.date.isoweekday()
col = dfjob.apply(fn, axis=1)
dfjob = dfjob.assign(date_day=col.values)

dfjob.drop(['active','created','time','duration','title','date','skills'], axis=1, inplace=True)
dfjob = pd.get_dummies(dfjob, columns=['shift','area','grade','departmentId','managerId','created_day','date_day'])

print(dfjob.shape)
pd.set_option('display.max_columns', None)
dfjob.head()
```

```python
# A nurse can either apply ('applied') for a job or skip it ('skipped').
# Than the manager will choose someone and send an offer ('confirmed') or reject the application of a nurse ('rejected').
# And now the nurse can choose to accept the job ('deal') or decline it ('rejectedByNurse').
# DELETE 'skipped'
# dfapp = dfapp[dfapp.status != 'skipped']
# dfapp.drop(['id','staffId', 'jobDate','offeredDate','status', 'active'], axis=1, inplace=True)
# dfapp = dfapp.groupby(['jobId']).size().reset_index(name='counts')

df = pd.merge(dfjob, dfapp, how='outer', left_on='id', right_on='jobId')
df = df[pd.isnull(df.id) == False]
df['counts'].fillna(0, inplace=True)

Y = pd.DataFrame(df['counts'])
fn = lambda row: 0 if (row.counts <= 2) else 1
col = Y.apply(fn, axis=1)
Y = Y.assign(y=col.values)
Y.drop(['counts'], axis=1, inplace=True)

df.drop(['id','confirmedApplicaitonId','offeredDate','status ','jobId','counts'], axis=1, inplace=True)
# Y = pd.concat([dfapp[dfapp['jobId'] == row['id']] for (index, row) in dfjob.iterrows()], ignore_index=True)

# dfjob.rename(index=str, columns={"id": "jobId"})
# dfjob.drop(['id','confirmedApplicaitonId','offeredDate','status '], axis=1, inplace=True)
# dfcount.reset_index(drop=True, inplace=True)
# Y.drop(['jobId'], axis=1, inplace=True)
# df = pd.concat([dfjob, dfcount], axis=1)
print(list(df))
print(df.shape)
print(Y.shape)
pd.set_option('display.max_columns', None)
print(df.head())
```

```python
file_name = './dataset.csv'

df.to_csv(file_name, encoding='utf-8',index=False)
Y.to_csv('./y.csv', encoding='utf-8',index=False)
```
